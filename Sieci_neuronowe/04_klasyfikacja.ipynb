```python
#BINARY CLASSIFICATION
#Input layer - same as number of features
#Hidden layers - depends on problem - min=1, max=unlimited
#Neurons per hidden layer - usually 10 to 100
#Output layer - 1 (one class or the other)
#Hidden activation function - usually ReLU
#Output activation function - sigmoid
#Loss functiion - cross entropy (tf.kreas.losses.BinaryCrossentropy)
#Optimizer - SGD, Adam
```

```python
#MULTICLASS CLASSIFICATION
#Input layer - as above
#Hidden laters - as above
#Neurons per hidden layer - as above
#Output layer - number of class
#Hidden activation function - as above
#Output activation function - softmax
#Loss functiion - cross entropy (tf.keras.losses.CategoricalCrossentropy)
#Optimizer - as above
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.BinaryCrossentropy(), optimizer=tf.keras.optimizers.Adam(), metrics='accuracy')

model.fit(X, Y, epochs = 200, verbose = 0)
model.evaluate(X,Y)
```

```python
def plot_decision_boundary(model, X, Y):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1  #tworzenie granic
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100)) #tworzenie siatki
    
    x_in = np.c_[xx.ravel(), yy.ravel()] #np.c_   konkatenacja list
    
    y_pred = model.predict(x_in)
    
    if len(y_pred[0]) > 1:
        print("doing multiclass classification")
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("doing binary classifiaction")
        y_pred = np.round(y_pred).reshape(xx.shape)
        
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

```python
#finding the best learning rate

#to find the ideal learning rate (the learning rate where the loss decreases the most during training)
#we are going to use the following steps:
# A learning rate callback
# Another model
# A modified loss curves plot
```

```python
tf.random.set_seed(42)

model_evo1 = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_evo1.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), metrics='accuracy')

#creating a learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))

history = model_evo1.fit(X_train, Y_train, epochs = 100, callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history.history).plot(figsize=(10,7), xlabel='epochs')
```

```python
#value of ideal learning rate is between 38 and 45 epochs
```
